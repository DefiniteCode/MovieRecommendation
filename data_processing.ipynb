```python
# create a copy of the dataframe as the dataframe and filter for only released movies
working_df = df[df["status"] == "Released"]
working_df.head()
```

```python
# mwe have some movies with release_date in the future so lets filter them as well
working_df = working_df[working_df["release_date"] <= "2025-04-22"].sort_values(by="release_date", ascending=False)
```

```python
working_df.head()
```

```python
# create a tags column which is a combination of genres and keywords -- remove commas and convert to lower 
working_df["tags"] = (df.genres + df.keywords).str.lower().str.replace(",","")
working_df.head()
```

```python
# creating a much smaller subset for the recommendation system to reduce load on computing resources
reduced_df = working_df.iloc[:10000]
reduced_df.head()
```

```python
# creating another subset for training th cosine similarity model
model_df = reduced_df[["id","title","tags"]].reset_index()
model_df.head()
```

```python
# drop the original index column
model_df.drop(columns=['index'], inplace=True)
```

```python
model_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#initialize vectorizer
tfidf = TfidfVectorizer(stop_words="english")

# transform tags to vectors
tfidf_matrix= tfidf.fit_transform(model_df["tags"])
```

```python
similarities = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
# function for recommend button
def get_recommendations(title):
    # Ensure exact match and safe indexing
    match = model_df[model_df['title'].str.lower() == title.lower()]
    if match.empty:
        return pd.DataFrame()

    index = match.index[0]

    similarity_scores = similarities[index]
    movie_scores = list(enumerate(similarity_scores))
    sorted_scores = sorted(movie_scores, key=lambda x: x[1], reverse=True)
    top_movies = sorted_scores[1:11]
    top_indices = [i for i, score in top_movies]

    return model_df.iloc[top_indices] 
```

```python
recommendations_df = get_recommendations("Sunburst")
recommendations_df
```
